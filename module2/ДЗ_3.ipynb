# Завдання 6
Перевірте чи правильно розраховано загальну вартість покупки. Виведіть ті рядки де є помилка. Виправте дані в таблиці

```python
filtered_items = df["sale_amt"]!=(df["units"]*df["unit_price"])

df.loc[filtered_items,"sale_amt"]= df["units"] * df["unit_price"]
```
